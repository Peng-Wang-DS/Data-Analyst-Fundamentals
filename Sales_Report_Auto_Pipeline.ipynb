Import Data from Kaggle for this demo and prepare for demo data

```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None) 

df = pd.read_csv('Balaji Fast Food Sales.csv')
df1=df.copy()
df2=df.copy()
df1.head()
Mapping=df1[['item_name','item_type','item_price']].drop_duplicates(subset='item_name').reset_index(drop=True)
data = df2[['item_name','date','time_of_sale','quantity','transaction_amount']]
data['date'] = data['date'].apply(pd.to_datetime, dayfirst=True)
data.sort_values(by='date')
data.reset_index(drop=True)    
data.head()
data1 = data.melt(
    id_vars = [ 'item_name','date'],
    value_vars= ['quantity','transaction_amount'],
    var_name= 'measure',
    value_name = 'value'
)

Data = data1.pivot_table(
    index=[ 'item_name','measure'],
    columns='date',
    values='value',
    aggfunc='sum',
    fill_value=0
)
```

```python
import pandas as pd
DATA = Data.copy().reset_index()
DATA.head(1)
```

```python
DATA2 = DATA.melt(
    id_vars=['item_name','measure'],
    value_vars = DATA.columns[2:],
    var_name = 'date',
    value_name = 'value'
)

DATA2['date'] = DATA2['date'].apply(pd.to_datetime,dayfirst=True).dt.normalize()

DATA2.head()
```

```python
DATA3 = DATA2.merge(Mapping, on='item_name', how ='left')
Raw_data = DATA3[['item_type',
                  'item_name',
                  'item_price',
                  'measure',
                  'date',
                  'value'
                  ]].sort_values(['item_type', 'item_name', 'date'])
Raw_data.head()
```

```python
with pd.ExcelWriter('Test.xlsx', engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    Raw_data.to_excel(writer, sheet_name='Raw_Data', index=False)

print("Data written to the 'Raw_Data' sheet successfully.")
```
