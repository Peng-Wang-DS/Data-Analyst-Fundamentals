Import Data from Kaggle for this demo and prepare for demo data

```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None) 

df = pd.read_csv('Balaji Fast Food Sales.csv')
df1=df.copy()
df2=df.copy()
df1.head()
Mapping=df1[['item_name','item_type','item_price']].drop_duplicates(subset='item_name').reset_index(drop=True)
data = df2[['item_name','date','time_of_sale','quantity','transaction_amount']]
data['date'] = data['date'].apply(pd.to_datetime, dayfirst=True)
data.sort_values(by='date')
data.reset_index(drop=True)    
data.head()
data['date']=data['date'].dt.date
data1 = data.melt(
    id_vars = [ 'item_name','date'],
    value_vars= ['quantity','transaction_amount'],
    var_name= 'measure',
    value_name = 'value'
)

Data = data1.pivot_table(
    index=[ 'item_name','measure'],
    columns='date',
    values='value',
    aggfunc='sum',
    fill_value=0
)

Data.reset_index(inplace=True)

```

```python
DATA2 = Data.melt(
    id_vars = ['item_name','measure'],
    value_vars = Data.columns[2:],
    value_name = 'value',
    var_name = 'date'
)

DATA2.dtypes
```

```python
DATA2['date'] = DATA2['date'].apply(pd.to_datetime, dayfirst=True)
DATA2.dtypes
```

```python
Raw_Data = DATA2.merge(Mapping, on='item_name', how='inner')

# display(
#     DATA2.shape,
#     Raw_Data.shape,
#     Raw_Data.head()
# )

Raw_Data = Raw_Data[['item_type','item_name', 'item_price','measure','date','value']]

Raw_Data.head(2)
```

```python
path = 'Sales Report Automation - Python.xlsx'

with pd.ExcelWriter(path, engine = 'openpyxl', mode='a', if_sheet_exists= 'replace') as writer:
    Raw_Data.to_excel(writer, sheet_name='Raw_Data', index=False)

print("Data written successfully to 'Raw_Data' sheet! ")
```
